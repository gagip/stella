```python
import cv2
import numpy as np
import requests
from average_hash_matching import *
```

```python
url = 'https://www.dlm1947.com/data/goods/2022/product/6I75481_detail_01.jpg'
image_nparray = np.asarray(bytearray(requests.get(url).content), dtype=np.uint8)
image: np.ndarray = cv2.imdecode(image_nparray, cv2.IMREAD_COLOR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
height, width = image.shape[:2] 

cv2.imshow('dd', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
is_all_white = np.sum(255 - image, 1) == 0
is_all_white_index = np.where(is_all_white)
is_all_white_index
d = np.argwhere(np.diff(is_all_white_index) > 1)[:,1]

result = []
result.append((0, is_all_white_index[0][0] - 1))
for idx, value in enumerate(d.tolist()):
    prev = is_all_white_index[0][value] + 1
    white_index = is_all_white_index[0][value+1] -1
    
    if white_index - prev > 50:
        result.append((prev, white_index))    
    
    
print(result)
# is_all_white_index[0]
# np.where(np.sum(image, 1) == 0)
```

```python
for r in result:
    start = r[0]
    end = r[1]
    
    cv2.imshow(f"{start}", image[start:end,:])
    cv2.waitKey(0)
cv2.destroyAllWindows()
```
